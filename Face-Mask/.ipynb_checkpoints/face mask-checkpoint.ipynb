```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.layers import Dense, Flatten, Conv2D
from tensorflow.keras import Sequential
from tensorflow import keras
from tensorflow.keras import layers
import os
import cv2
```

```python
categories = os.listdir('data')
```

```python
categories
```

```python
data =[]
for cat in categories:
  counter = 0
  for file in os.listdir(f'data/{cat}'):
    if counter<=500:
      label = categories.index(cat)
    # img_path = f'{file}'
      img = cv2.imread(f'data/{cat}/{file}')
      img = cv2.resize(img,(224,224))
      data.append([img,label])
    else:break
    counter+=1
```

```python
import random
random.shuffle(data)
X = []
y = []

for img,label in data:
  X.append(img)
  y.append(label)
```

```python
X = np.array(X)
y = np.array(y)
X = X/255
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
from keras.applications.vgg16 import VGG16
vgg = VGG16()
# # vgg.summary()
```

```python
model = tf.keras.Sequential([
    layers.Conv2D(16, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(256, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(32, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(16, activation="leaky_relu"),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')
])
```

```python
vgg_model = Sequential()
for i in vgg.layers[:-1]:
  vgg_model.add(i)
vgg_model.add(Dense(1,activation = 'sigmoid'))
```

```python
vgg_model.summary()
```

```python
vgg_model.compile(loss= 'binary_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
```

```python
vgg_model.fit(X_train,y_train,epochs = 10,validation_data = (X_test,y_test))
```

```python
def preprocess_image(frame):
    frame = cv2.resize(frame, (224, 224))  # Resize to match model input
    frame = frame.astype('float32') / 255.0  # Normalize pixel values
    frame = np.expand_dims(frame, axis=0)  # Add batch dimension (1, 224, 224, 3)
    return frame
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()  # Capture frame-by-frame
    if not ret:
        break

    # # Preprocess the frame for prediction
    processed_frame = preprocess_image(frame)

    # # Make prediction
    prediction = model.predict(processed_frame)
    predicted_class = 'mask on' if prediction[0] > 0.5 else 'Please wear mask'  # Adjust for your classes

    # # Display the prediction on the video feed
    if predicted_class == 0:
        cv2.
        
    cv2.putText(frame, f"Prediction: {predicted_class}", (50, 50),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow('Live Camera Feed', frame)  # Show the frame with prediction

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
```
